```python
df.drop(columns=["diffgr:id", "msdata:rowOrder", "Arrival_Date"], inplace=True)

print(df.head(), '\n')
print(df.info())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

temp = df.iloc[:, 0:6]
for col in temp.columns:
    df[col] = label_encoder.fit_transform(df[col])
        
print(df.info(), '\n')
print(df.describe())
```

```python
X = df.iloc[:, [1, 2, 5, 6, 7]].values.reshape(-1, 5)
y = df.iloc[:, -1].values.reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
X = standard_scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_b = np.c_[np.ones((len(X_train), 1)), X_train]

# Gradient Descent parameters
learning_rate = 0.001
n_iterations = 100000
m = len(X_b)  # number of samples

# Initialize weights randomly
np.random.seed(23)
theta = np.random.randn(X_train.shape[1] + 1, 1)  # theta is (2, 1) because we have 2 parameters (bias + 1 feature)

# Gradient Descent
eps = 1e-3

# while True:
#     gradients = 1/m * X_b.T.dot(X_b.dot(theta) - y_train)  # Compute the gradient
#     theta = theta - learning_rate * gradients  # Update weights
#     if (max(np.abs(gradients)) < eps): break

for i in range(n_iterations):    
    gradients = 1/m * X_b.T.dot(X_b.dot(theta) - y_train)  # Compute the gradient
    theta = theta - learning_rate * gradients  # Update weights    
```

```python
# Output the final parameters
print("Theta (final parameters):\n", theta)
```

```python
# Make predictions
X_test_b = np.c_[np.ones((len(X_test), 1)), X_test]  # Add the bias term
y_pred1 = X_test_b.dot(theta)  # Predict using the trained model
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred1)
mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred1)

print("Gradient Descent in LR")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared: {r2}")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred2 = model.predict(X_test) 
```

```python
mae = mean_absolute_error(y_test, y_pred2)
mse = mean_squared_error(y_test, y_pred2)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred2)

print("Linear Regression Model")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared: {r2}")
```
